```python
#importing dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

```python
#loading a dataset

df = pd.read_csv("/content/insurance.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
encode = LabelEncoder()
```

```python
df['sex'] = encode.fit_transform(df['sex'])
df['smoker'] = encode.fit_transform(df['smoker'])
df['region'] = encode.fit_transform(df['region'])
```

```python
df.head()
```

```python
x = df.drop(['charges'], axis = 1)
y = df['charges']
```

```python
#standardizing the data
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
y
```

```python
#splitting data into train and test

x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#loading a model

model = LinearRegression()
```

```python
#training part
x_trained = model.fit(x_train,y_train)
```

```python
#evaluating a model on training data

y_pred = model.predict(x_train)
```

```python
#R2 score

r2_score = metrics.r2_score(y_train, y_pred)
print("The R2 score is : ", r2_score )
```

```python
#evaluating a model on testing data
y_pred = model.predict(x_test)
r2_score = metrics.r2_score(y_test, y_pred)
print("The R2 score is : ", r2_score )
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cv_scores = cross_val_score(model, x, y, cv=5, scoring='r2')

print("Cross-validation R2 scores:", cv_scores)
print("Mean R2 score:", cv_scores.mean())
print("Standard deviation of R2 scores:", cv_scores.std())
```

Now let's perform cross-validation with a Decision Tree Regressor and compare its performance to the Linear Regression model.

```python
from sklearn.tree import DecisionTreeRegressor

# Instantiate a Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=2)

# Perform 5-fold cross-validation on the Decision Tree model
dt_cv_scores = cross_val_score(dt_model, x, y, cv=5, scoring='r2')

print("Decision Tree Cross-validation R2 scores:", dt_cv_scores)
print("Decision Tree Mean R2 score:", dt_cv_scores.mean())
print("Decision Tree Standard deviation of R2 scores:", dt_cv_scores.std())

print("\nComparison:")
print(f"Linear Regression Mean R2 score: {cv_scores.mean():.4f}")
print(f"Decision Tree Mean R2 score: {dt_cv_scores.mean():.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate a Random Forest Regressor
rf_model = RandomForestRegressor(random_state=2)

# Perform 5-fold cross-validation on the Random Forest model
rf_cv_scores = cross_val_score(rf_model, x, y, cv=5, scoring='r2')

print("Random Forest Cross-validation R2 scores:", rf_cv_scores)
print("Random Forest Mean R2 score:", rf_cv_scores.mean())
print("Random Forest Standard deviation of R2 scores:", rf_cv_scores.std())

print("\nComparison:")
print(f"Linear Regression Mean R2 score: {cv_scores.mean():.4f}")
print(f"Decision Tree Mean R2 score: {dt_cv_scores.mean():.4f}")
print(f"Random Forest Mean R2 score: {rf_cv_scores.mean():.4f}")
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='r2')
```

```python
grid_search.fit(x, y)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best R2 Score:", best_score)
```

```python
# Instantiate a new Random Forest Regressor with the best hyperparameters
best_rf_model = RandomForestRegressor(**best_params, random_state=2)

# Train the new model using the training data
best_rf_model.fit(x_train, y_train)

# Make predictions on the test data
y_pred_best_rf = best_rf_model.predict(x_test)

# Calculate and print the R2 score of the model's predictions on the test set
r2_score_best_rf = metrics.r2_score(y_test, y_pred_best_rf)
print("R2 score of the Random Forest model with best hyperparameters on the test set:", r2_score_best_rf)
```
